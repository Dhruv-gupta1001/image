```python
import matplotlib,cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
def adjust_gamma(image, gamma=1.0):
        invGamma = 1.0 / gamma
 
        table = np.array([((i / 255.0) ** invGamma) * 255
                for i in np.arange(0, 256)]).astype("uint8")
        return cv2.LUT(image, table)
```

```python
original = cv2.imread('cointry2.png',cv2.IMREAD_COLOR)
```

```python
plt.imshow(cv2.cvtColor(original, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
for gamma in np.arange(0.0, 3.5, 0.5):
        if gamma == 1:
                continue
        gamma = gamma if gamma > 0 else 0.1
        adjusted = adjust_gamma(original, gamma=gamma)
```

```python
plt.imshow(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB))
coins_thres=(cv2.cvtColor(adjusted, cv2.COLOR_BGR2RGB))
cv2.imwrite('coinsf.png',coins_thres)
```

```python
img =cv2.imread('coinsf.png',cv2.IMREAD_COLOR)
plt.title('Enhanced Image')
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
